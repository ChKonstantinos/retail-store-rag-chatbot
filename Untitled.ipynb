```python
from src.config import RAW_DATA_DIR, CHUNK_SIZE, CHUNK_OVERLAP
from src.document_loader import load_documents
from src.text_splitter import split_documents
from src.vector_store import create_vector_store

documents = load_documents(RAW_DATA_DIR)

chunks = split_documents(
    documents=documents,
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

vector_store = create_vector_store(chunks)

print("Vector store created successfully.")
print(f"Chunks stored: {len(chunks)}")
```

```python
from src.vector_store import load_vector_store

vector_store = load_vector_store()

query = "What should I do if a customer wants to return a product without a receipt?"

results = vector_store.similarity_search(query, k=3)

for i, doc in enumerate(results, start=1):
    print("=" * 80)
    print(f"Result {i}")
    print("Source:", doc.metadata.get("title"))
    print("File:", doc.metadata.get("file_name"))
    print("Chunk:", doc.metadata.get("chunk_id"))
    print()
    print(doc.page_content)
```
